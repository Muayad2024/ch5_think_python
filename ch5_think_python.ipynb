```python
from time import time 
now = time()
minute = now // 60
hours = minute // 60
days = hours // 24
years = days // 365
print("Current minute:", minute)
print("Current hour:", hours)
print("Current day:", days)
print("Current year:", years)
```

```python
def is_triangle(x, y, z):
    if x + y > z or x + z > y or y + z > x:
        print ("Yes")
    elif x + y == z or x + z == y or y + z == x:
        print ("Degenerate")
    else:
        print('no')
```

```python
def recurse(n, s):
    if n == 0:
        print(s)
    else:
        recurse(n-1, n+s)
recurse(3, 0)
```

```python
from jupyturtle import forward, left, right, back, make_turtle
def draw(length):
    angle = 50
    factor = 0.6
    if length > 5:
        forward(length) 
        left(angle)
        draw(factor * length) 
        right(2 * angle)
        draw(factor * length) 
        left(angle)
        back(length)
make_turtle(width=500, height=400, delay=0.2) 
draw(30)
```
